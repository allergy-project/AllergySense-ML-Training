```python
import cv2
import os
from PIL import Image
count = 0
```

Video To Imgae

```python
path = "kulit-sehat"
save_path = 'foto-kulit-sehat' 
files = os.listdir(path)
print(files)
for file in files:
    print(file)
    vidcap = cv2.VideoCapture("kulit-sehat/" + file)
    success, image = vidcap.read()
    while success:
        cv2.imwrite(save_path + "/frame%d.jpg" % count, image)  # save frame as JPEG file
        success, image = vidcap.read()
        print("Read a new frame: ", success)
        count += 1
    else:
        print("not success")
```

Crop Image in Center

```python
path = "raw-data-penyakit-kulit/train/Acne/"
save_path = 'dataku-kulit-gasehat-acak' 
files = os.listdir(path)
crop_shape = (150, 150)
print(files)



for file in files:
    # Open the image using PIL
    image = Image.open(path + '/' + file)

    # Get the width and height of the image
    width, height = image.size

    # Calculate the number of quadrants that can fit in the image
    num_quadrants_width = width // crop_shape[0]
    num_quadrants_height = height // crop_shape[1]

    # Calculate the offset to center the crop within each quadrant
    offset_x = (crop_shape[0] - crop_shape[0] // 2) // 2
    offset_y = (crop_shape[1] - crop_shape[1] // 2) // 2

    # Crop each quadrant and save the resulting images
    for i in range(num_quadrants_height):
        for j in range(num_quadrants_width):
            left = j * crop_shape[0] + offset_x
            top = i * crop_shape[1] + offset_y
            right = left + crop_shape[0]
            bottom = top + crop_shape[1]

            quadrant = (left, top, right, bottom)
            cropped_image = image.crop(quadrant)
            cropped_image.save(save_path + f'/kulit-sakit{count}.jpg')
            count += 1
```
